```python
from os.path import join
import pandas as pd
```

```python
# get all the 1.1.3.15 identifiers
df = pd.read_csv('../final/brenda_2017_1/ec_uid_org_from_fasta_2017_1.tsv', sep='\t')
ec_identifiers = set(df[df.ec == '1.1.3.15'].uid.values)
ec_identifiers
```

```python
# first selection id's from all enzyme classes
with open(join('first_selection', 'selected_ids.tsv')) as f:
    first_sel = set(f.read().split('\n'))

first_sel = first_sel & ec_identifiers
len(first_sel)
```

```python
# second selection id's from all enzyme classes
with open(join('second_selection', 'selected_ids.tsv'), 'r') as f:
    second_sel = set(f.read().split('\n'))

second_sel = second_sel & ec_identifiers
len(second_sel)
```

```python
74+59
```

```python
not_synthesized = set(['A0A0H6MLS9',
                'A0A0U5PE87',
                'Q5WFZ5',
                'A0A0U3IJJ5',
                'Q46IL6',
                'Q8YI46',
                'Q2LXM8',
                'W6R6M4',
                'A0A0W0SKN3',
                'A0A0M2HFS4',
                'H0RVQ4'])


all_sel = first_sel | second_sel
all_sel = all_sel - not_synthesized
len(all_sel) # these are the 122 used in the paper
```

```python
all_sel = sorted(list(all_sel))
with open('selected_ids.tsv', 'w') as f:
    f.write('\n'.join(all_sel))
```

```python
with open('sequence_order.tsv', 'w') as f:
    with open(join('first_selection', 'sequence_order.tsv'), 'r') as fi:
        f.write(fi.readline())
        
    for uid in all_sel:
        with open(join('first_selection', 'sequence_order.tsv'), 'r') as fi:
            for line in fi:
                if uid in line:
                    f.write(line)
            
        with open(join('second_selection', 'sequence_order.tsv'), 'r') as fi:
            for line in fi:
                if uid in line:
                    f.write(line)
            
```

```python
with open(join('experiments', 'detailed', '1-1-3-15_batch1_summary.tsv'), 'w') as f:
    with open(join('experiments', 'detailed', '1-1-3-n_batch1_summary.tsv'), 'r') as fi:
        f.write(fi.readline())

    for uid in all_sel:
        with open(join('experiments', 'detailed', '1-1-3-n_batch1_summary.tsv'), 'r') as fi:
            for line in fi:
                if uid in line:
                    f.write(line)

with open(join('experiments', 'detailed', '1-1-3-15_batch2_summary.tsv'), 'w') as f:
    with open(join('experiments', 'detailed', '1-1-3-n_batch2_summary.tsv'), 'r') as fi:
        f.write(fi.readline())

    for uid in all_sel:
        with open(join('experiments', 'detailed', '1-1-3-n_batch2_summary.tsv'), 'r') as fi:
            for line in fi:
                if uid in line:
                    f.write(line)
          
                  
```
