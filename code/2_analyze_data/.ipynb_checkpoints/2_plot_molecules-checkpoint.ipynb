```python
from cheminformatics import cheminfo

# print out the conda or miniconda env
import subprocess
result = subprocess.run(['conda', 'env', 'export', '--no-builds'], stdout=subprocess.PIPE)
conda_env = '\n'.join(result.stdout.decode().strip().split('\n')[:-1])
display(print(conda_env))
```

#### Make the conversion from name to smile

```python
mols = ['Lactate', 'Glycolate', '2-hydroxyglutarate', 'mandelate', '2-hydroxystearate', '2-hydroxyoctanoate']

chem_obj = cheminfo.NameToSmile(names=mols, retest_none=False)
```

#### Convert smiles to a data object

```python
data_obj = cheminfo.SmileToData(names=chem_obj.names(),
                                smiles=chem_obj.smiles(), 
                                fingerprint='layered', 
                                metric='tanimoto')

```

#### Draw the structures as svg

```python
img = data_obj.draw_structures(highlight_substructure=True, use_svg=True)

display(img)

with open(join(FIGURES, 'structures.svg'), 'w') as f:
    f.write(img.data)
```
