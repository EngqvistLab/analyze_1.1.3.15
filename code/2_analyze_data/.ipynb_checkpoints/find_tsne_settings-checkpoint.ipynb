#### Load the 1.1.3.15 sequences encoded using Moran autocorrelation

```python
# got a bunch of data from the iFeatures server, let's plot them

## load autocorrelation data and drop duplicates ##
filepath = join(FINAL, 'BRENDA', 'Moran.tsv')
moran = pd.read_csv(filepath, sep='\t').drop_duplicates()

# replace the first column with identifiers
moran['#'] = [i.split(';')[0] for i in moran['#']]
moran.rename(columns={'#':'uid'}, inplace=True)
#display(moran.head())


## load uid to ec mapping and merge ##
filepath = join(FINAL, 'brenda_data_2019_1', 'ec_uid_org_from_fasta.tsv')
uid_ec = pd.read_csv(filepath, sep='\t').drop_duplicates()
#display(uid_ec.head())

moran = moran.merge(uid_ec, on=['uid'])


## load organism and temperature information and merge ##
filepath = join(FINAL, 'brenda_data_2019_1', '1-1-3-n_identifier_info.tsv')
uid_property = pd.read_csv(filepath, sep='\t').drop_duplicates()

uid_property.drop(['lineage_identifiers', 'lineage_ranks', 'lineage_names'], 
                  inplace=True,
                 axis=1)
#display(uid_property.head())

moran = moran.merge(uid_property, on=['uid'])


# # only keep 1.1.3.15
moran_subset = moran[moran['ec']=='1.1.3.15']
# moran_subset = moran

display(moran_subset.head())
display(moran_subset.describe())

```

#### I also want to get data regarding which cluster the sequences are in

```python
uid_dict = {k:None for k in moran_subset.uid.values}

filepath = join(FINAL, 'BRENDA')

for fi in os.listdir(filepath):
    if fi.endswith('sequences.fasta'):
        cluster = fi.replace('_sequences.fasta', '')
        
        with open(join(filepath, fi), 'r') as f:
            for line in f:
                if line.startswith('>'):
                    uid = line.strip('>').split(';')[0]
                    uid_dict[uid] = cluster
                    
data = {'uid':[], 'cluster':[]}
for k,v in uid_dict.items():
    data['uid'].append(k)
    data['cluster'].append(v)
            

moran_subset = moran_subset.merge(pd.DataFrame(data), on='uid')
    
display(moran_subset.head())
display(moran_subset.tail())
```

#### I need to load the activity data

```python
filepath = join(FINAL, 'brenda_data_2019_1')

batch1 = pd.read_csv(join(filepath, '1-1-3-n_batch1_summary.tsv'), sep='\t')
batch2 = pd.read_csv(join(filepath, '1-1-3-n_batch2_summary.tsv'), sep='\t')

# rename the column for the identifier
batch1.rename(columns={'node':'uid'}, inplace=True)
batch2.rename(columns={'name':'uid'}, inplace=True)

# add info on which batch it is
batch1['batch'] = 'first'
batch2['batch'] = 'second'

# check that there are no duplicates
for uid in batch1.uid.values:
    if uid in batch2.uid.values:
        print('Bad overlap, uid %s exists in both datasets' % uid)
    

# combine them 
both_batches = batch1.append(batch2).drop_duplicates()

# drop unnessecary columns
both_batches = both_batches[(both_batches['synthesized'] == 'Yes') |
                            (both_batches['synthesized'] == 'No')]
both_batches.drop(['rep1_soluble', 
                   'rep1_active', 
                   'rep2_soluble', 
                   'rep2_active', 
                   'rep3_soluble', 
                   'rep3_active'], axis=1, inplace=True)
    
display(both_batches.head())


# make sure all the identfiers are there in the main dataset
# for uid in both_batches.uid.values:
#     assert uid in moran_subset.uid.values
# print('all uids present in main dataset')

    
# now merge with the other data
moran_subset = moran_subset.merge(both_batches, on='uid', how='left')

# replace nans
moran_subset = moran_subset.replace(np.nan, '', regex=True)

display(moran_subset.head())
display(moran_subset.tail())
```

#### Make a t-sne visualization

```python

def make_plot(plot_data, name):
    '''
    '''
    plt.scatter(plot_data[:,0], plot_data[:,1], s=2, color='red')
    plt.title(name)
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.tight_layout()  


X = PCA(n_components=50).fit_transform(moran_data)

# scale the pca to use as initialization
PCAinit = X[:,:2] / np.std(X[:,0]) * 0.0001

# calculate tsne using pca initialization
my_plot = fast_tsne(moran_data, initialization=PCAinit)


display(make_plot(my_plot, 't-SNE, 1.1.3.15'))
```

```python
def highlight_part_of_plot(plot_data, selected_data, background_data):
    '''
    Plot a single ec number
    '''

    plt.scatter(plot_data[:,0][background_data], plot_data[:,1][background_data], s=7, color=[0.8, 0.8, 0.8])
    plt.scatter(plot_data[:,0][selected_data], plot_data[:,1][selected_data], s=7, color='red')
    plt.title('t-SNE')
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.tight_layout()
    
    
def highlight_by_color(plot_data, colors):
    '''
    Color t-sne plot by categorical values
    '''
    plt.scatter(plot_data[:,0], plot_data[:,1], s=7, color=colors)
        
    plt.title('t-SNE')
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.legend()
    plt.tight_layout()
    
    
def get_colors_for_items(item_list):
    '''
    Get a set of unique colors
    '''
    item_set = sorted(list(set(item_list)))
    
    colors = scales.rainbow(len(item_set))
    
    return {k:v for k,v in zip(list(item_set), colors)}



def make_ordered_color_list(input_list, input_dict):
    '''
    Make a list of colors that signify some information in the fasta header
    '''
    data = []
    for thing in input_list:
        
        data.append(str(input_dict[thing]))
        
    return np.array(data)
```

#### Color t-SNE by clusters to validate that it is good

```python
# get a list of all clusters
items_list = moran_subset.cluster.values

# get colors
items_dict = get_colors_for_items(items_list)
color_list = make_ordered_color_list(items_list, items_dict)

# plot it 
display(highlight_by_color(my_plot, color_list))


set(items_list)
color_list
```
