## Open questions
Do I use natural substrates or all substrates?
Should I factor in Km and Kcat?

Separate between promiscuous EC classes and those artificially appearing promiscuous because they have been jumbled together?
    Very difficult to do since few enzymes have been comprehensively tested
    Also, data is connected to organism rather than sequence in most cases
        
        
We hypothesize that phylogeny can help
    For a promiscuous EC class we expect enzymes from the same or closely related species to be active on several dissimilar substrates
    For an artificially promiscuous EC class we expect enzymes from the same or closely related species to be active on only very similar substrates. Put another way, we expect idividual substrates to be associated with limited phylogenetic diversity
    
How do I differentiate between one and several enzymes from an organism?
    
How does one validate this?
    Can I use the EC class from the bachelors?

#### First I need to get a list of all the substrate product pairs

```python
import json
import re
from cheminformatics import helpfunctions, cheminfo, brenda
import random
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns; sns.set(style="ticks", color_codes=True)


def get_natural_substrate_pairs(filepath):
    """
    Process the BRENDA data and retain the first molecule 
    named as the main substrate and the first molecule of the products.
    Restricted to the natural substrates category.
    """

    with open(filepath, 'r') as f:
        in_data = json.loads(f.read())

    out_data = set([])
    for ec in sorted(in_data.keys()):
        
        for entry in in_data[ec].split('!'):
            for item in entry.split('#'):
                #print(item)

                if item.startswith(u'naturalReactionPartners'):
                    reaction = item.replace(u'naturalReactionPartners*', u'')
                    
                    # first attempt to get all substrates and all products separated
                    try:
                        reactants, products = reaction.split('=')
                    except:
                        reactants = reaction
                        products = ''

                    # split the reactants
                    try:
                        reactant_list = reactants.split(' + ')
                    except:
                        reactant_list = reactants
                        
                    # take the first one
                    substrate = re.sub('^[0-9]+?[ ]', '', reactant_list[0].strip())
                    
                                                              
                    # split the products
                    try:
                        product_list = products.split(' + ')
                    except:
                        product_list = products
                        
                    # take the first one
                    product = re.sub('^[0-9]+?[ ]', '', product_list[0].strip())
                    
                    if substrate == 'more' or product == 'more':
                        continue
                        
                    # clean the name and standardize it
                    substrate = helpfunctions.clean_name(substrate)
                    product = helpfunctions.clean_name(product)
                        
                    out_data.add((substrate, product))
                                                    
                
    return sorted(list(out_data))
    
    

def get_substrate_pairs(filepath):
    """
    Process the BRENDA data and retain the first molecule 
    named as the main substrate and the first molecule of the products.
    Includes all substrates listed.
    """

    with open(filepath, 'r') as f:
        in_data = json.loads(f.read())

    out_data = set([])
    for ec in sorted(in_data.keys()):
        
        for entry in in_data[ec].split('!'):
            for item in entry.split('#'):
                #print(item)

                if item.startswith(u'reactionPartners'):
                    reaction = item.replace(u'reactionPartners*', u'')
                    
                    # first attempt to get all substrates and all products separated
                    try:
                        reactants, products = reaction.split('=')
                    except:
                        reactants = reaction
                        products = ''

                    # split the reactants
                    try:
                        reactant_list = reactants.split(' + ')
                    except:
                        reactant_list = reactants
                        
                    # take the first one
                    substrate = re.sub('^[0-9]+?[ ]', '', reactant_list[0].strip())
                    
                                                              
                    # split the products
                    try:
                        product_list = products.split(' + ')
                    except:
                        product_list = products
                        
                    # take the first one
                    product = re.sub('^[0-9]+?[ ]', '', product_list[0].strip())
                    
                    if substrate == 'more' or product == 'more':
                        continue
                        
                    # clean the name and standardize it
                    substrate = helpfunctions.clean_name(substrate)
                    product = helpfunctions.clean_name(product)
                        
                    out_data.add((substrate, product))
                                                    
                
    return sorted(list(out_data))
    
    
filepath = join(RAW_EXTERNAL, 'brenda_data_2019_1', 'natural_substrates_data.txt')
pairs = get_natural_substrate_pairs(filepath)

filepath = join(RAW_EXTERNAL, 'brenda_data_2019_1', 'substrates_data.txt')
pairs = get_substrate_pairs(filepath)
```

#### Now I need to get the smiles (where possible) for each substrate and product

```python
smile_pairs = []
name_pairs = []

for p in pairs:

    chem_obj = cheminfo.NameToSmile(names=list(p), retest_none=False)
        
    smiles = chem_obj.smiles(exclude_none=False)
    if None in smiles or smiles == []:
        continue
        
    else:
        smile_pairs.append(smiles)
        name_pairs.append(list(p))


        
print('Starting pairs: %s' % len(pairs))
print('Pairs with smiles: %s' % len(smile_pairs))
print('Percentage left: %s%%' % round(100*len(smile_pairs)/len(pairs)))
```

#### Go through each pair and compute similarity, also test with random substrate

```python
from os.path import exists, join

# read the saved file if it exists, to save computation
filepath = join(FINAL, folder_name, 'substate_product_sims.tsv')
if exists(filepath):
    df = pd.read_csv(filepath, sep='\t')

else:
    fingerprints = ['maccs', 'pattern', 'layered', 'morgan2', 'rdkit']
    measures = ['asymmetric', 'braunblanquet', 'cosine', 'dice', 'kulczynski',  'mcconnaughey',  'rogotgoldberg', 'russel', 'sokal', 'tanimoto']
    num_pairs = len(name_pairs)

    print('To process: %s' % num_pairs)
    counter = 0
    random.seed(47)
    data = {'fingerprint':[], 'measure':[], 'comparison':[], 'substrate':[], 'product':[], 'random':[], 'similarity':[]}
    for s, n in zip(smile_pairs, name_pairs):

        # keep track of progress
        counter += 1
        if counter % 250 == 0:
            print('Completed: %s' % counter)
        
        # get a random product
        index = random.randint(0, num_pairs-1)
        random_name = name_pairs[index][1]
        random_smile = smile_pairs[index][1]
        
        # I get a SMILE parse error for mesoheme
        if n[0] == 'mesoheme'or n[1] == 'mesoheme' or random_name == 'mesoheme':
            continue
      
        # now calculate similarity using different fingerprints an measures
        for f in fingerprints:
            for m in measures:

                # compute similarity for substrate with own product
                data_obj = cheminfo.SmileToData(names=n,
                                            smiles=s, 
                                            fingerprint=f, 
                                            metric=m)

                # add to data structure
                data['fingerprint'].append(f)
                data['measure'].append(m)
                data['comparison'].append('substrate_product')
                data['substrate'].append(n[0])
                data['product'].append(n[1])
                data['random'].append(random_name)
                data['similarity'].append(data_obj.similarity().values[0, 1])


                # compute similarity for substrate with random product
                data_obj = cheminfo.SmileToData(names=[n[0], random_name],
                                            smiles=[s[0], random_smile], 
                                            fingerprint=f, 
                                            metric=m)

                # add to data structure
                data['fingerprint'].append(f)
                data['measure'].append(m)
                data['comparison'].append('substrate_random')
                data['substrate'].append(n[0])
                data['product'].append(n[1])
                data['random'].append(random_name)
                data['similarity'].append(data_obj.similarity().values[0, 1])


    # convert to data frame
    df = pd.DataFrame(data)
    
    # save to disk
    df.to_csv(filepath, sep='\t', index=False)
```

#### Get the delta of the mean and plot, I want to find the fingerprint with the greatest delta

I ended up picking layered with Tanomoto for further analysis as this gives the greatest delta. Mcconnaughey is technically larger but the scale of that measure is twice as large, so the score should be devided in half.

```python
# split up the substrate-product comparisons and the substrate-random ones
df_sp = df[df['comparison'] == 'substrate_product']
df_sr = df[df['comparison'] == 'substrate_random']

# rename the similarity mean column
df_sp.rename(columns={'similarity':'similarity_sp'}, inplace=True)
df_sr.rename(columns={'similarity':'similarity_sr'}, inplace=True)

# transfer the substrate_random mean column to the other data frame
df_combined = df_sp.merge(df_sr, on=['fingerprint', 'measure', 'substrate', 'product', 'random'])

# subtract one from the other
df_combined['diff_mean'] = df_combined['similarity_sp'] - df_combined['similarity_sr']

# calculate the mean and standard deviation for each fingerprint and measure pair 
df_delta = pd.DataFrame(df_combined.groupby(['fingerprint', 'measure']).mean()).reset_index()
df_delta['diff_sd'] = pd.DataFrame(df_combined.groupby(['fingerprint', 'measure']).std()).reset_index().diff_mean

df_delta


```

#### Keep only the summary stats, and devide the Mcconnaughey fingerprint data by 2 to fix the scale

```python
# first drop the original similarity values
df_delta_mod = df_delta.drop('similarity_sp', axis=1)
df_delta_mod = df_delta_mod.drop('similarity_sr', axis=1)

display(df_delta_mod[df_delta_mod['measure']=='mcconnaughey'])

# now divide the mcconnaughey measures by half
df_delta_mod.loc[df_delta_mod['measure']=='mcconnaughey', 'diff_mean'] = df_delta_mod[df_delta_mod['measure']=='mcconnaughey']['diff_mean'] / 2
df_delta_mod.loc[df_delta_mod['measure']=='mcconnaughey', 'diff_sd'] = df_delta_mod[df_delta_mod['measure']=='mcconnaughey']['diff_sd'] / 2

display(df_delta_mod[df_delta_mod['measure']=='mcconnaughey'])

# sort on the mean difference
df_delta_mod.sort_values(by=['diff_mean'], inplace=True, ascending=False)

# make a new column with both fingerprint and measure
df_delta_mod['fingerprint_and_measure'] = df_delta_mod['fingerprint'] + ' + ' + df_delta_mod['measure']

display(df_delta_mod)
```

#### Using the best fingerprint, go through each enzyme class and compute substrate similarity

```python

# read the saved file if it exists, to save computation
filepath = join(FINAL, folder_name, 'ec_class_sims.tsv')
if exists(filepath):
    df_ec = pd.read_csv(filepath, sep='\t')
    
else:
    mol_obj = brenda.BrendaNaturalMols(typeof='substrate') 

    # a list of all ec numbers
    ec_nums = sorted(mol_obj.ec())

    # a dictionary with ec number keys holding a list of substrates
    mol_data = mol_obj.data_dict()

    # go through each ec and compute similarity measures
    data = {'ec':[], 'subst_count':[], 'max':[], 'min':[], 'mean':[], 'median':[], 'stdev':[]}
    for ec in ec_nums:

        # get object that converts names to smiles
        chem_obj = cheminfo.NameToSmile(names=mol_data[ec], retest_none=False)

        # get substrates and smiles
        names = chem_obj.names(exclude_none=True)
        smiles = chem_obj.smiles(exclude_none=True)
        
        # I get a SMILE parse error for mesoheme, so I need to remove it
        if 'mesoheme' in names:
            ix = names.index('mesoheme')
            del names[ix]
            del smiles[ix]
        
        # skip if only one substrate
        num_substrates = len(names)
        if num_substrates < 2:
            continue
        
        # use the smiles to get a data object that can be used to compute similarity
        data_obj = cheminfo.SmileToData(names=names,
                                smiles=smiles, 
                                fingerprint='layered', 
                                metric='tanimoto')

        # get global similarity data
        sim_data = data_obj.global_similarity_stats()

        # add to data structure
        data['ec'].append(ec)
        data['subst_count'].append(num_substrates)
        data['max'].append(sim_data['max'])
        data['min'].append(sim_data['min'])
        data['mean'].append(sim_data['mean'])
        data['median'].append(sim_data['median'])
        data['stdev'].append(sim_data['stdev'])    
        
    # convert to data frame
    df_ec = pd.DataFrame(data)
    
    # save to disk
    df_ec.to_csv(filepath, sep='\t', index=False)
    
```

#### Pull in data regarding how many sequences are in each of the enzyme classes

```python

def count_seqs_in_ec_when_clustered_together(ec_list):
    '''
    Count how many cluster representatives are in each ec class.
    '''
    # first make a list with all the uniprot identifiers from the cluster representatives
    cluster_path = join(INTERMEDIATE, 'brenda_data_2019_1', 'clustering', 'clustered_sequences_70.fasta')
    
    cluster_uids = set([])
    with open(cluster_path, 'r') as f:
        for line in f:
            
            if line.startswith('>'):
                uid = line.strip('>').split(';')[0]
                cluster_uids.add(uid)
                
                
    # now go through the ec fasta files and see where they fall
    filepath = join(RAW_EXTERNAL, 'brenda_data_2019_1', 'sequence_data')
    
    print('%s files to process' % len(ec_list))
    ec_count = {'ec':[], 'seq_count':[]}
    counter = 0
    for ec in ec_list:
        
        counter += 1
        if counter % 100 == 0:
            print('%s files analyzed' % counter)
        
        count_in_ec = 0
        if exists(join(filepath, '%s.fasta' % ec)):
            with open(join(filepath, '%s.fasta' % ec), 'r') as f:
                for line in f:
                    if line.startswith('>'):
                        uid = line.strip('>').split(';')[0]

                        if uid in cluster_uids:
                            count_in_ec += 1
        
        # add to data structure
        ec_count['ec'].append(ec)
        ec_count['seq_count'].append(count_in_ec)
                        
    return pd.DataFrame(ec_count)



def count_seqs_in_ec_when_clustered_individually(ec_list):
    '''
    Count how many cluster representatives are in each ec class.
    '''
  
    # now go through the ec fasta files and see where they fall
    filepath = join(INTERMEDIATE, 'brenda_data_2019_1', 'individual_ec_clustering')
    
    print('%s files to process' % len(ec_list))
    ec_count = {'ec':[], 'seq_count':[]}
    counter = 0
    for ec in ec_list:
        
        counter += 1
        if counter % 100 == 0:
            print('%s files analyzed' % counter)
        
        count_in_ec = 0
        ec_filepath = join(filepath, '%s_clustered_sequences_40.fasta' % ec)
        if exists(ec_filepath):
            with open(ec_filepath, 'r') as f:
                for line in f:
                    if line.startswith('>'):
                        count_in_ec += 1
        
        # add to data structure
        ec_count['ec'].append(ec)
        ec_count['seq_count'].append(count_in_ec)
                        
    return pd.DataFrame(ec_count)


identity_cutoff = 40
filepath = join(FINAL, folder_name, 'ec_class_seq_counts_%s.tsv' % (identity_cutoff))
if exists(filepath):
    df_counts = pd.read_csv(filepath, sep='\t')
    
else:
    df_counts = count_seqs_in_ec_when_clustered_individually(ec_list=df_ec.ec.values)
    df_counts.to_csv(filepath, sep='\t', index=False)
    
df_counts.head()
```

```python

# get a list of all the selected identifiers 
selected = moran_subset[(moran_subset.synthesized == 'No') | (moran_subset.synthesized == 'Yes')]

# get a list with all the identifiers not selected
not_selected = moran_subset[moran_subset.synthesized == '']

# load up the identity data
df_ident = pd.read_csv(join(FINAL, 'similarity_data', 'all_identity_matrix_ident.tsv'), 
                       sep='\t', 
                      index_col=0)

display(df_ident.head())


# make sure all the identifiers are present in the identity matrix
for uid in selected.uid.values:
    assert uid in df_ident.index, 'Error, %s not present in identity matrix' % uid

for uid in not_selected.uid.values:
    assert uid in df_ident.index, 'Error, %s not present in identity matrix' % uid

    
# make a subset of the identity matrix
df_ident_selected = df_ident.loc[selected.uid.values]


sim_data = {'uid':[], 'max_sim':[]}
for uid in moran_subset.uid.values:
    sim_data['uid'].append(uid)
    sim_data['max_sim'].append(max(df_ident_selected[uid].values))

sim_data_df = pd.DataFrame(sim_data)
display(sim_data_df.head())

# merge with the other data
moran_subset = moran_subset.merge(sim_data_df, on=['uid'])
display(moran_subset.head())
```

#### Make a t-sne visualization

```python
import numpy as np
import sklearn
from sklearn.decomposition import PCA
from sklearn.manifold import MDS

from colcol import scales

# import tsne
import sys; sys.path.append('/home/localadmin/github/FIt-SNE')
from fast_tsne import fast_tsne


def make_plot(plot_data, name):
    '''
    '''
    plt.scatter(plot_data[:,0], plot_data[:,1], s=2, color='red')
    plt.title(name)
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.tight_layout()  


X = PCA(n_components=50).fit_transform(moran_data)

# scale the pca to use as initialization
PCAinit = X[:,:2] / np.std(X[:,0]) * 0.0001

# calculate tsne using pca initialization
my_plot = fast_tsne(moran_data, initialization=PCAinit)


display(make_plot(my_plot, 't-SNE, 1.1.3.15'))
```

```python
def highlight_part_of_plot(plot_data, selected_data, background_data):
    '''
    Plot a single ec number
    '''

    plt.scatter(plot_data[:,0][background_data], plot_data[:,1][background_data], s=7, color=[0.8, 0.8, 0.8])
    plt.scatter(plot_data[:,0][selected_data], plot_data[:,1][selected_data], s=7, color='red')
    plt.title('t-SNE')
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.tight_layout()
    
    
def highlight_by_color(plot_data, colors):
    '''
    Color t-sne plot by categorical values
    '''
    plt.scatter(plot_data[:,0], plot_data[:,1], s=7, color=colors)
        
    plt.title('t-SNE')
    plt.gca().get_xaxis().set_ticklabels([])
    plt.gca().get_yaxis().set_ticklabels([])
    plt.legend()
    plt.tight_layout()
    
    
def get_colors_for_items(item_list):
    '''
    Get a set of unique colors
    '''
    item_set = sorted(list(set(item_list)))
    
    colors = scales.rainbow(len(item_set))
    
    return {k:v for k,v in zip(list(item_set), colors)}



def make_ordered_color_list(input_list, input_dict):
    '''
    Make a list of colors that signify some information in the fasta header
    '''
    data = []
    for thing in input_list:
        
        data.append(str(input_dict[thing]))
        
    return np.array(data)
```
